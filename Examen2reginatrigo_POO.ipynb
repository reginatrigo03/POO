# Una clase es una plantilla o un plano que define la estructura y comportamiento de un objeto, son de suma importancia debido a los siguientes puntos:  
1.   Abstracción: abstraer los detalles complejos de un objeto y ofrecen una interfaz clara para interactuar con ese objeto

1.   Reutilización: la reutilización de código
2.   Encapsulación: los datos y métodos relacionados deben estar contenidos dentro de una clase
1.   Herencia: una clase herede propiedades y métodos de otra clase
2.   Polimorfismo: objetos de diferentes clases respondan de manera similar











```python
#Creamos una clase llamada trigo, con argumentos (nombre, guess_who y promedio)
class Trigo():

    def __init__(self):
        print("La clase Trigo se ha creado")

    def guess_who(self):
        print("Yo soy Trigo")

    def promedio(self):
        print("Llevo un promedio de 95 en la clase de progrmación agosto-diciembre 2023")
```

```python
# Cursor
trigo = Trigo()
trigo
```

```python
trigo.guess_who()
```

```python
trigo.promedio()
```

```python
# Creamos ahora una clase con polimorfismos para "Trigo"
class Trigo_bcn():

    def __init__(self, name, promedio, faltas):
        self.name = name
        self.promedio = promedio
        self.faltas = faltas

    def nombre(self):
        return self.name + "soy Trigo"

    def goles(self):
        return self.promedio + "95"

    def campeonatos(self):
        return self.faltas + "0 en total"
```

**Ejercicio 4:** A partir de un diccionario, crea un DataFrame (que llamarás **precios_muestra**) que consista de dos columnas y que tenga 12 valores **flotantes** que eligirás de manera aleatoria en el intervalo (8,13). Etiqueta del 1 al 12 y llama a la columna de etiquetas Mes.

1.- Construye una función que calcule los rendimientos simples de cada columna.

2.- Las desviaciones estadísticas de los rendimientos ayudan a calcular la volatilidad del portafolio:
$$\text{desviaciones}=(\text{rendimientos}-\text{media}(\text{rendimientos}))^2.$$
Crea una función que calcule la media de las desviaciones y luego la volatilidad de tu portafolio (raíz cuadrada de la media de las desviaciones).

3.- Anualizamos la volatilidad escalándola (multiplicándola) por la raíz cuadrada del número de períodos por observación.

Por lo tanto, para anualizar la volatilidad de una serie mensual, la multiplicamos por la raíz cuadrada de 12. Muestra la serie de la volatilidad anualizada.

4.- Realiza una gráfica de barras para los rendimientos y una gráfica en líneas para los precios.

5.- Calcula los rendimientos compuestos para cada columna y después anualiza los rendimientos.

---

```python
# Creamos el data frame con números aleatoreos
# Usamos el random.uniform para poner números flotantes
# Utilizamos el range para dicirle que se indexen ambas columnas
import numpy as np
import random
np.random.seed(1)
import pandas as pd

precios_muestra={'Mes':list(range(1,13)),'Precio':[random.uniform(8,13)for i in range(12)]}
```

```python
df=pd.DataFrame(precios_muestra)
df
```

```python
rendimientos=df.pct_change
```

```python
#Calculamos rendimientos
rendimientos= df['Precio'].pct_change().dropna()
rendimientos
```

```python
media_rend=rendimientos.mean()
```

```python
# Calculamos la desviación estándar de los rendimientos
desv = (rendimientos-media_rend)**2
desv
```

```python
# Calculamos la media de los rendimientos
media=desv.mean()
media
```

```python
# Anualizamos
volat=desv*(1/12)
volat
```

```python
volat=media*(1/12)
volat
```

```python
# Generamos gráfica
rendimientos.hist(bins=10)
```

```python
#importamos módulos para la gráfica
import matplotlib.pyplot as plt
from pylab import plot
import seaborn as sns
```

```python
df['Precio'].plot(label="Precio",figsize=(10,5))
```

```python
# Calculamos los rendimientos compuestos
rendimientos_compuestos=(1+rendimientos).cumprod()
rendimientos_compuestos
```
